```python
#How much companies spent on the R & D ?
print("Company's spending for R & D in $8N")
RD = df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000
plt.show()
```

```python
#Draw a Bar plot to show the amount spent on R & D by the companies

plt.bar(RD.index, RD.values, color=['blue', 'black', 'green'])
plt.title('R&D Spending by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_%8n')

plt.show()

```

```python
df.head()
```

```python
print("Company'S AI_Revenue_USD_Mn")

rev = df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000

rev
```

```python
# Draw a Bar Plot to show the Revenues of the companies

plt.bar(rev.index, rev.values, color=['blue', 'black', 'green'], width = 0.3)

plt.title("Revenues of the companies")
plt.xlabel("company")
plt.ylabel("Amount in the USD_$8n")

plt.show()
```

```python
# Bar plot to show the expenditure & revenue of the companies

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)

plt.bar(RD.index, RD.values, color=['blue', 'black', 'green'])
plt.title('R&D Spending by the companies')
plt.xlabel('Company')
plt.ylabel('Amount in USD_%8n')

plt.subplot(1,2,2)

plt.bar(rev.index, rev.values, color=['blue', 'black', 'green'], width = 0.3)
plt.title("Revenues of the companies")
plt.xlabel("company")
plt.ylabel("Amount in the USD_$8n")

plt.show()
```
